##### 1   Basic Encapsulation (Private Variable)

Create a class Person with:

A private variable __salary
A method to set and get the salary

```python
class Person:
    def __init__(self,name,salary):
        self.name=name #public variable
        self.__salary=salary #private variable

    def get_salary(self):
        return self.__salary

    def set_salary(self,new_salary):
        self.__salary+=new_salary    

p=Person("Himanshu",450000)
print(p.name)
print(p.get_salary())

p.set_salary(500000)
print(f"Updated Salary:",p.get_salary())
```

##### 2.  Getter and Setter Methods

Create a class Student with a private attribute __marks.
Implement:

set_marks() → to assign value
get_marks() → to return value

```python
class Student:
    def __init__(self):
        self.__marks=0

    #Set Method
    def set_marks(self,marks):
        if 0<= marks <=100:
            self.__marks=marks
        else:
            print("Invalid! Marks should be greater than zero and less than 100")
    
    #Get Method
    def get_marks(self):
        return self.__marks

s=Student()

s.set_marks(85)
print("Marks:",s.get_marks())

s.set_marks(120) # Invalid Choice
```

##### 3.  Encapsulation with Validation

Create a class BankAccount with:

Private variable __balance
Method deposit(amount)
Method withdraw(amount) (should check for insufficient balance)

```python
class BankAccount:
    def __init__(self,initial_balance=0):
        self.__balance=initial_balance

    def deposit(self,amount):
        self.__balance+=amount
        print(f"Deposited: {amount} | Current Balance: {self.__balance}")

    def withdraw(self,amount):
        if self.__balance<amount:
            print(f"Insufficient Funds: {self.__balance}") 
        else:
            self.__balance-=amount
            print(f"Withdrawal: {amount} | Current Balance: {self.__balance}")

    def get_balance(self):
        return self.__balance

a1=BankAccount()
a1.deposit(5000)
a1.withdraw(3000)
a1.withdraw(2500)
```

##### 4. Using Property Decorators

Create a class Employee with a private variable __salary.
Use:

@property to access salary
@salary.setter to update salary with validation

```python
class Employee:
    def __init__(self,salary):
        self.__salary = salary

    # getter using property
    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self,value):
        if value > 0:
            self.__salary=value
        else:
            print("Invalid salary")    

e1=Employee(50000)
print("Salary:", e1.salary)

#Updated Salary
e1.salary=450000
print("Updated Salary: ",e1.salary)

e1.salary=-450 # Invalid Choice
```
