##### 1.  Method Overriding (Runtime Polymorphism)

Create a base class Shape with a method area().
Create subclasses Circle and Rectangle that override the area() method.

```python
class Shape:
    def area(self):
        print("Area of a shape")

class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius

    def area(self):
        return 3.14 * self.radius * self.radius

class Rectangle(Shape):
    def __init__(self,width,length):
        self.width=width
        self.length=length

    def area(self):
        return self.length * self.width

def result(shape):
    print(f"Area : {shape.area()}")


result(Circle(2))
result(Rectangle(3,9))
```

##### 3.  Operator Overloading

Create a class Point with attributes x and y.
Overload the + operator to add two objects of this class.

```python
class Point:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __add__(self,other):
        return Point (self.x + other.x , self.y + other.y)

    def __str__(self):
        return f"({self.x},{self.y})"

# Created Objects
p1 = Point(2,3)
p2 = Point(8,9)

# Add points
p3 = p1 + p2

print(f"Result: {p3}")
```

##### 4.  Method Overloading (Python Style)

Python doesn’t support traditional method overloading.
Write a class Calculator where a method add() can handle:

Two numbers
Three numbers

```python
class Calculator:
    def add(self, *args):
        return sum(args)

cal = Calculator()   

print(f"Two Numbers: {cal.add(7,9)}")
print(f"Three Numbers: {cal.add(8,7,9)}")
```
