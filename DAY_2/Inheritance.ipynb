##### 1.  Single Inheritance (Basic)

Create a class Animal with a method sound().
Create a child class Dog that inherits from Animal and overrides the sound() method.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        print("Animal makes a sound")


class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)

    def sound(self):  # overriding
        print(f"{self.name} says Woof")


# Object
d1 = Dog("Roxy")
d1.sound()
```

##### 2.  Multilevel Inheritance

Create three classes:

Vehicle → method start()
Car (inherits Vehicle) → method drive()
ElectricCar (inherits Car) → method charge()

Create an object of ElectricCar and call all methods.

```python
class Vehicle:
    def start(self):
        print("Vehicle started")


class Car(Vehicle):
    def drive(self):
        print("Car is driving")


class ElectricCar(Car):
    def charge(self):
        print("Electric car is charging")

e1 = ElectricCar()

e1.start()    # from Vehicle
e1.drive()    # from Car
e1.charge()   # from ElectricCar
```

##### 3.  Multiple Inheritance

Create two classes:

Father → method skills()
Mother → method talents()

Create a child class Child that inherits from both and accesses both methods.

```python
class Father:
    def skills(self):
        print("Father's skills: Driving, Coding")


class Mother:
    def talents(self):
        print("Mother's talents: Cooking, Teaching")


class Child(Father, Mother):
    pass

c1 = Child()

# Access both methods
c1.skills()    # from Father
c1.talents()   # from Mother
```
