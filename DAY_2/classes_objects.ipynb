##### 1.  Create a Simple Class

Create a class Student with:

Attributes: name, age, marks
Method: display_details() to print student info

```python
class Student():
    def __init__(self,name,age,marks):
        self.name = name
        self.age = age
        self.marks = marks

    def display_details(self):
        print("------Student Details-------")
        print(f"Name: " ,self.name)
        print(f"Age: " ,self.age)
        print(f"Marks: " ,self.marks)
        print()

s1 = Student("Himanshu",22,45)   
s2 = Student("Srikar", 22, 46)
s3 = Student("Avanish",22, 47)

s1.display_details()
s2.display_details()
s3.display_details()
```

##### 2.  Constructor & Instance Variables

Write a class Car that:

Initializes brand, model, and price using a constructor (__init__)
Has a method to display car details

```python
class Car():
    def __init__(self,brand,model,price):
        self.brand=brand
        self.model=model
        self.price=price

    def display_details(self):
        print("-----Car Details------")
        print("Brand Name: ", self.brand)
        print("Model number: ", self.model)
        print("Car Price: ", self.price)
        print()

c1= Car("AUDI-VIRTUS", 25678, 2000000)
c2= Car("TATA-SIERRA", 89764, 1149000)

c1.display_details()
c2.display_details()
```

##### 3.  Class vs Instance Variable

Create a class Employee with:

Class variable: company_name
Instance variables: name, salary


```python
class Employee():
    company_name = "GalaxEye Pvt Ltd"

    def __init__(self,name,salary):
        self.name = name
        self.salary = salary

    def display_details(self):
        print("-------Company Details------")
        print(f"Company Name: {Employee.company_name}")
        print(f"Employee Name: {self.name}")
        print(f"Employee Salary: {self.salary}")
        print()

e1=Employee("Himanshu Goyal",60000)
e2=Employee("Srikar Kulkarni",660000)

e1.display_details()
e2.display_details()
```

##### 4.  Basic OOP Problem (Bank Account)

Create a class BankAccount with:

Attributes: account_holder, balance
Methods:
deposit(amount)
withdraw(amount) (check for insufficient balance)
display_balance()

```python
class BankAccount():
    bank_Name="State Bank of India (SBI)"

    def __init__ (self,account_holder,balance):
        self.account_holder=account_holder
        self.balance=balance

    def deposit(self,amount):
        self.balance+=amount
        print(f"{BankAccount.bank_Name}")
        print(f"{amount} is Deposited, New Balance is {self.balance}")

    def withdraw(self,amount):
        if amount>self.balance:
            print(f"Insuffient Balance, Current Balance is {self.balance}")
        else:
            self.balance-=amount
            print(f"{amount} is withdraw, Updated Balance is {self.balance}")    

    def get_balance(self):
        return self.balance     


a1 = BankAccount("Himanshu", 5000)    
a1.deposit(300)
a1.withdraw(600)
a1.get_balance()
```
