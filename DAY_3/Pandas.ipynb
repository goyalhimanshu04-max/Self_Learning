##### 2.  Filtering Data

Given a DataFrame with student data (name, marks, age):

Extract students with marks greater than 75
Extract students whose age is less than 18

```python
import pandas as pd

data = {
    "Name":["Saransh","Aniket","Rushikesh","Gaurav","Srikar"],
    "Age":[21,18,19,20,17],
    "Marks":[91,82,97,65,77]
}

df=pd.DataFrame(data)

## Highest Marks
highest_mks = df[df['Marks']>85]
print("Highest Marks:  \n",highest_mks)

## Young Students
young_age = df[df["Age"]<18]
print("\n Young Student: \n",young_age)
```

##### 3.  Handling Missing Values

Given a DataFrame with missing values:

Detect missing values
Fill them with a specific value (e.g., 0 or mean)
Drop rows with missing values

```python
import pandas as pd

data = {
    "Name":["Saransh","Aniket","Rushikesh",None,"Srikar"],
    "Age":[21,18,None,20,None],
    "Marks":[91,None,97,None,77]
}

df=pd.DataFrame(data)

# 1.Detect Missing Values
print("Missing Values\n",df.isnull())
print("\nSum of Missing Values\n",df.isnull().sum())

# 2. Filling Missing values
filled_zero = df.fillna(0)
df_filled_mean = df.fillna(df.mean(numeric_only=True))

print("\nFilled with Zero\n",filled_zero)
print("\nFIlled with Mean\n",df_filled_mean)

# 3. Drop Missing Rows
dropp_rows = df.dropna()
print("\n Dropped rows\n",dropp_rows)
```

##### 4.  GroupBy Operation

Given a DataFrame with columns Department and Salary:

Find the average salary per department
Count number of employees in each department

```python
import pandas as pd

data = {
    "Name": ["Aman", "Riya", "Karan", "Neha", "Rahul", "Sneha"],
    "Department": ["IT", "HR", "IT", "HR", "Finance", "IT"],
    "Salary": [60000, 50000, 70000, 52000, 65000, 72000]
}

df = pd.DataFrame(data)

#1. Avg salary per department
Avg_salary = df.groupby("Department")["Salary"].mean()

#2. Count num of employess per department
num_of_emp = df.groupby("Department")["Name"].count()

print("Average Salary\n", Avg_salary)
print("\nCount of Employees\n", num_of_emp)
```
