##### 1. Distribution Plot

Write a program to:

Plot a distribution (histogram + KDE) of a numerical column
Customize bins and color

```python
import seaborn as sns
import matplotlib.pyplot as plt

values=[100,60,40,80,90]

sns.histplot(values,bins=2,edgecolor="black",color="skyblue",kde=True)
plt.title("Category Analysis")
plt.xlabel("values")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

##### 2. Box Plot for Outliers

Given a dataset:

Create a box plot to detect outliers
Compare distributions across categories

```python
import seaborn as sns
import matplotlib.pyplot as plt

department=["IT", "IT", "IT", "HR", "HR", "HR", "Finance", "Finance", "Finance"]
salary=[50000, 52000, 150000, 40000, 42000, 41000, 60000, 61000, 62000]

sns.boxplot(x=department,y=salary,color="blue")
plt.title("Analysis compare to Days")
plt.xlabel("Days")
plt.ylabel("Revenue")
plt.show()

```

##### 3.  Count Plot (Categorical Data)

Write a program to:

Plot the count of each category in a column (e.g., Gender, Department)

```python
import seaborn as sns
import matplotlib.pyplot as plt

department=["IT", "HR", "IT", "Finance", "HR", "IT", "Finance", "HR"]

sns.countplot(x=department,color="orange")
plt.title("Employes count by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

##### 4. Heatmap (Correlation Matrix)

Write a program to:

Compute correlation of a dataset
Visualize it using a heatmap with annotations

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data={
    "Age":[22, 25, 30, 35, 40],
    "Salary":[25000, 30000, 40000, 50000, 60000],
    "Experience":[1, 3, 5, 7, 10]
}

df=pd.DataFrame(data)

corr=df.corr()
corr

sns.heatmap(corr,annot=True,cmap="coolwarm")

plt.title("Correlation")
plt.show()
```

##### 5. Pair Plot (Multivariate Analysis)

Write a program to:

Create a pair plot of a dataset
Use a categorical column as hue

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    "Age": [22, 25, 30, 35, 40, 28, 32, 38],
    "Salary": [25000, 30000, 40000, 50000, 60000, 35000, 45000, 55000],
    "Experience": [1, 3, 5, 7, 10, 4, 6, 8],
    "Department": ["IT", "HR", "IT", "Finance", "Finance", "HR", "IT", "HR"]
}

df = pd.DataFrame(data)

sns.pairplot(df, hue="Department")
plt.show()
```
