##### 1. Sum of List Element

```python
def sum_elements(list):
    return sum(list)

print("Sum of List:",sum_elements([2,3,5,6]))

```

##### 2. Largest Element of List

```python
def large_element(list):
    return max(list)

print("Largest Element:",large_element([44,78,92,22,94]))
```

##### 3. Remove Duplicate from List

```python
def duplicate_element(list):
    seen=set()
    result=[]

    for num in list:
        if num not in seen:
            seen.add(num)
            result.append(num)
    return result      

print("Duplicate Element:",duplicate_element([1,2,3,4,2,5,1,3]))      
```

##### 4. Check if all Element in a List are Unique

```python
def unique_element(list):
    return len(list) == len(set(list))

print("Unique Element:",unique_element([44,54,32]))
```

##### 5. Reverse a List

```python
def rev_list(list):
    result=[]
    for i in range(len(list)-1,-1,-1):
        result.append(list[i])
    return result    

print("Reversed List:",rev_list([11,22,33,44,55,66,77,88,99,100]))
```

##### 6. Count num of ODD and EVEN Element in List:

```python
def count_odd_even(list):
    even = sum(1 for i in list if i%2==0)
    odd = sum(1 for i in list if i%2!=0)
    return (even,odd)

print("Count of Odd and Even: ",count_odd_even([22,13,42,32,28,43]))
```

##### 7. Max Difference between Consecutive Elements

```python
def max_consecutive(lst):
    max_diff=0

    for i in range(len(lst)-1):
        diff = abs(lst[i]- lst[i+1])
        if diff>max_diff:
            max_diff=diff
    return max_diff        

print("Maximum Consecutive: ",max_consecutive([10,45,44,33,55,42]))
```

##### 8. Merge Two Sorted List

##### 9. Rotate a String

```python
def rotate_string(lst, k):
    n=len(lst)

    if n == 0:
        return lst
    
    k=k%n

    for _ in range(k):
        last = lst[-1]

        for i in range(n-1,0,-1):
            lst[i]=lst[i-1]
        lst[0]=last
    return lst       

print("Rotate a String: ",rotate_string([2,3,4,5],3)) 
```

##### 10. Merge two list into Dictonary (Method 1)

```python
def merge_list_to_dict(keys,values):
    result={}

    for i in range(min(len(keys),len(values))):
        result[keys[i]]=values[i]
    return result    

print("Merge List: ",merge_list_to_dict(["a","b","c"],[1,2,3]))
```

```python
def merge_list(keys,values):
    return zip(keys,values)

print("Merge List:",dict(merge_list(["x","y","z"],[33,44,55])))
```

##### 11. Merge Multiple Dictionaries (Method 1)

```python
def merge_dict(dict1,dict2,dict3):
    result={}

    for d in (dict1,dict2,dict3):
        for key in d:
            result[key]=d[key]
    return result

print("Merging 3 Dictionaries:",merge_dict( {"a": 1},{"b": 2},{"c": 3}))    
```

##### 12. Word Frquency in Sentence

```python
def word_freq(sent):
    words=sent.split()

    result={}

    for word in words:
        if word in result:
            result[word]+=1
        else:
            result[word]=1
    return result            

print("Word Count:",word_freq("Run milkha Run"))
```

##### 13. Palindrome Tuple

```python
def palindrome_tuple(tup):
    n=len(tup)

    for i in range(n//2):
        if tup[i]!=tup[n-i-1]:
            return False
    return True

print("Palindrome:",palindrome_tuple([1,2,3,2,1]))    
print("Palindrome:",palindrome_tuple([1,2,3,4,1]))    
```

##### 14. Merge Dictionary with common dictionary

```python
def merge_dict(dicts):
    result={}

    for d in dicts:
        for key in d:
            if key in result:
                result[key]+=d[key]
            else:
                result[key]=d[key]
    return result

data = [
    {"a": 10, "b": 20},
    {"b": 5, "c": 15},
    {"a": 2, "c": 3}
]

print("Merged Dictionary:", merge_dict(data))
            
```

##### 15. Check if List is subset of another list

```python
def is_subset(lst1,lst2):
    for i in range(len(lst1)):
        found = False
        for j in range(len(lst2)):
            if lst1[i] == lst2[j]:
                found=True
                break
        if not found:
            return False
    return True

print("Is Subset:",is_subset([1, 2], [1, 2, 3, 4]))
```
