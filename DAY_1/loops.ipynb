```python
# 1. Print Pattern (Right Triangle)

rows = int(input("Enter the number of rows: "))

for i in range(1, rows+1):
    for j in range(i):
        print("+", end=" ")
    print()        
```

```python
# 2. Sum of Digits

number = int(input("Enter numbers: "))
sum_digit = 0

while number > 0:
    digit = number % 10
    sum_digit += digit
    number = number // 10

print(sum_digit)
```

```python
# 3. Reverse a Number

number = int(input("Enter number: "))
reverse = 0

while number > 0:
    digit = number % 10
    reverse = reverse * 10 + digit
    number = number // 10

print(reverse)    
```

```python
# 4. Fibonacci Series

step = int(input("Enter the number of step: "))
a,b = 0,1

for i in range(step):
    print(a, end=" ")
    a,b = b,a+b
```

```python
# 5. Prime Number Check

num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(f"{num} is prime number")    
```
