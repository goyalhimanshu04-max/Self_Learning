```python
# 1. Square All Elements Using map()

numbers = [1,5,6,8]
square= list(map(lambda x: x**2,numbers))
print(square)
```

```python
# 2. Filter Even Numbers

numbers = [2,5,9,8]
even_num = list(filter(lambda x: x%2==0, numbers))
print(even_num)
```

```python
# 3. Convert Strings to Uppercase

words = ["library", "ai/ml", "mlops"]

upper_Case = list(map(str.upper,words))
print(upper_Case)
```

```python
# 4. Use lambda with sorted()

data = [(1, 3), (4, 1), (2, 2)]

sorted_list = sorted(data,key=lambda x:x[1])

print(sorted_list)

```
